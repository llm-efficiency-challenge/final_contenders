```python
fetch_task_json_path_bigbench = lambda task_name: f"https://raw.githubusercontent.com/google/BIG-bench/main/bigbench/benchmark_tasks/{task_name}/task.json"
```

```python
WRITE_DIR = "../../data/bigbench_relevant_tasks"
```

```python
for curr_task in bigbench_tasks:
    json_url = fetch_task_json_path_bigbench(curr_task)
    if len(bigbench_tasks[curr_task])==0:
        response = requests.get(json_url)
        data = json.loads(response.text)
        save_path = os.path.join(WRITE_DIR, f"{curr_task}.json")
        if not os.path.exists(save_path):
            with open(save_path, 'w') as fd:
                print("Saving for: ", curr_task)
                json.dump(data, fd, indent=2)
    else:
        for subtask in bigbench_tasks[curr_task]:
            json_url  = f"https://raw.githubusercontent.com/google/BIG-bench/main/bigbench/benchmark_tasks/{curr_task}/{subtask}/task.json"
            response = requests.get(json_url)
            data = json.loads(response.text)
            save_path = os.path.join(WRITE_DIR, f"{curr_task}|{subtask}.json")
            if not os.path.exists(save_path):
                with open(save_path, 'w') as fd:
                    print("Saving for: ", curr_task)
                    json.dump(data, fd, indent=2)
    #break
```

```python
df_arr = []
tmnt_id = 0

for _task in wanted_tasks:
    print("########")
    print("Task: ", _task)
    '''if "five" not in str( _task):
        continue'''
        
    task_obj = seqio.get_mixture_or_task(_task)
    ds = task_obj.get_dataset(
            sequence_length={"inputs": 5000, "targets": 5000},  #  Trim long token sequences.
            split="all",  # Available splits = "all", "train", "validation".
        )
    done = False
    
    for d in tfds.as_numpy(ds):
        eg_obj = dict()
        eg_obj['tmnt_id'] = tmnt_id
        tmnt_id+=1
        
        eg_obj['task_id'] = _task
        
        eg_obj['instruction'] = d['inputs_pretokenized'].decode("utf-8")
        if not done:
            print("QUERY: ", eg_obj['instruction'] )
            done = True
        #print("ANS: ", d['answers'][0].decode("utf-8"))
        if 'answers' in d:
            eg_obj['output'] = d['answers'][0].decode("utf-8")
        else:
            eg_obj['output'] = d['targets_pretokenized'].decode("utf-8")
        df_arr.append(eg_obj)
            
```

```python
import random
from collections import defaultdict

def split_array_conditioned_on_X(array, ratio, attribute):
    # Separate the array into lists based on the attribute value
    attribute_lists = defaultdict(list)
    for d in array:
        attribute_value = d[attribute]
        attribute_lists[attribute_value].append(d)

    # Calculate the split index for each list based on the ratio and split the lists
    split_lists = defaultdict(list)
    for attribute_value, attribute_list in attribute_lists.items():
        split_index = int(len(attribute_list) * ratio)
        random.shuffle(attribute_list)
        split_lists[attribute_value] = (attribute_list[:split_index], attribute_list[split_index:])

    # Combine the lists to form the final two arrays
    array_1 = []
    array_2 = []
    for attribute_value, (list_1, list_2) in split_lists.items():
        array_1.extend(list_1)
        array_2.extend(list_2)

    return array_1, array_2
```
