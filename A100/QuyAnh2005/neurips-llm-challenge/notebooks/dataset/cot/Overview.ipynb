```python
def get_inputs(sample):
    input_string = sample[0]
    # Split the string into two parts: the question and the options
    parts = input_string.split('\\nOptions:\\n')

    # Extract the question and options
    question = parts[0]
    options = parts[1]

    # Replace '\\n' with newline character '\n' for better formatting
    options = options.replace('\\n', '\n')
    options = options.replace("(", "").replace(")", ".")
    # options = [str(option[1:2]) + str(option[3:]) for option in optionss]

    # Combine them into the desired format
    formatted_string = f'Question: {question}\n{options}'

    return formatted_string

def get_response(sample):
    sample[2] = sample[2].replace('\\n', '\n')
    return f"Answer: {sample[1][1:-1]} \nExplanation: {sample[2]}"

print(aqua_df.apply(get_inputs, axis=1)[0])
print(aqua_df.apply(get_response, axis=1)[0])
```

# creak_train.tsv

```python
creak_df = pd.read_csv(names[2], sep="\t", header=None)
creak_df
```

```python
row = 3
creak_df.iloc[row, 0], creak_df.iloc[row, 1], creak_df.iloc[row, 2]
```

```python
def get_inputs(sample):
#     print(sample[0])
    input_string = sample[0]
    input_string = input_string.replace('\\n', '\n')
    try:
        # Split the string into the claim and options
        parts = input_string.split('\nOptions:\n')

        # Extract the claim and options
        claim = parts[0]
        options = parts[1]

        # Replace the hyphen with letters A and B
        options = options.replace('- yes', 'A. Yes')
        options = options.replace('- no', 'B. No')

        # Combine them into the desired format
        formatted_string = f'{claim}\n{options}'
        return formatted_string
    except Exception as e:
        return f"Question: {input_string} \nA. Yes \nB. No"

    return formatted_string

def get_response(sample):
    sample[1] = sample[1].strip().capitalize()
    sample[2] = sample[2].replace('\\n', '\n')
    if sample[1] == 'Yes':
        return f"Answer: A. {sample[1]} \nExplanation: {sample[2]}"
    if sample[1] == 'No':
        return f"Answer: B. {sample[1]} \nExplanation: {sample[2]}"


creak_df.apply(get_inputs, axis=1)[20]
# print(creak_df.apply(get_inputs, axis=1)[10])
# print(creak_df.apply(get_response, axis=1)[0])
```

```python
creak_df['inputs'] = creak_df.apply(get_inputs, axis=1)
creak_df['response'] = creak_df.apply(get_response, axis=1)

row = 17
print(creak_df.iloc[row, 3])
print(creak_df.iloc[row, 4])
```

# ecqa_train.tsv

```python
ecqa_df = pd.read_csv(names[3], sep="\t", header=None)
ecqa_df
```

```python
row = 17
ecqa_df.iloc[row, 0], ecqa_df.iloc[row, 1], ecqa_df.iloc[row, 2]
```

```python
def get_inputs(sample):
    input_string = sample[0]
    input_string = input_string.replace('\\n', '\n')
    # Split the string into the question and options
    parts = input_string.split('\nOptions:\n')

    # Extract the question and options
    question = parts[0]
    options = parts[1]

    # Replace hyphens with letters A, B, C, D, E
    options = options.split('\n')
    options = [option[2:] for option in options]
    formatted_options = '\n'.join([f"{chr(65 + i)}. {option}" for i, option in enumerate(options)])

    # Combine them into the desired format
    formatted_string = f'{question}\n{formatted_options}'
    return formatted_string

def get_response(sample):
    input_string = sample[0]
    input_string = input_string.replace('\\n', '\n')
    # Split the string into the question and options
    parts = input_string.split('\nOptions:\n')

    # Extract the question and options
    question = parts[0]
    options = parts[1]

    # Replace hyphens with letters A, B, C, D, E
    options = options.split('\n')
    options = [option[2:] for option in options]

    answer = sample[1]
    for i, option in enumerate(options):
        if option == answer:
            formatted_answer = f"{chr(65 + i)}. {option}"

    formatted_string = f"Answer: {formatted_answer} \nExplanation: {sample[2]}"
    return formatted_string

ecqa_df.apply(get_inputs, axis=1)[10]
ecqa_df.apply(get_response, axis=1)[10]
```

```python
ecqa_df['inputs'] = ecqa_df.apply(get_inputs, axis=1)
ecqa_df['response'] = ecqa_df.apply(get_response, axis=1)

row = 17
print(ecqa_df.iloc[row, 3])
print(ecqa_df.iloc[row, 4])
```

```python
def get_inputs(sample):
    input_string = sample[0]
    input_string = input_string.replace('\\n', '\n')
    try:
        # Split the string into the question and options
        parts = input_string.split('\nOptions:\n')

        # Extract the question and options
        question = parts[0]
        options = parts[1]

        # Replace hyphens with letters A, B, C, D, E
        options = options.split('\n')
        options = [option[2:] for option in options]
        formatted_options = '\n'.join([f"{chr(65 + i)}. {option.capitalize()}" for i, option in enumerate(options)])

        # Combine them into the desired format
        formatted_string = f'{question}\n{formatted_options}'
        return formatted_string
    except Exception as e:
        formatted_string = f"{input_string} \nA. Yes \nB. No \nC. It is not possible to tell"
        return formatted_string

def get_response(sample):
    input_string = sample[0]
    input_string = input_string.replace('\\n', '\n')
    try:
        # Split the string into the question and options
        parts = input_string.split('\nOptions:\n')

        # Extract the question and options
        question = parts[0]
        options = parts[1]

        # Replace hyphens with letters A, B, C, D, E
        options = options.split('\n')
        options = [option[2:] for option in options]

        answer = sample[1]
        for i, option in enumerate(options):
            if option == answer:
                formatted_answer = f"{chr(65 + i)}. {option}"

        formatted_string = f"Answer: {formatted_answer} \nExplanation: {sample[2]}"
    except Exception as e:
        answer = sample[1]
        if answer == 'yes':
            formatted_answer = 'A. Yes'
        elif answer == 'no':
            formatted_answer = 'B. No'
        else:
            formatted_answer = 'C. It is not possible to tell'
    formatted_string = f"Answer: {formatted_answer} \nExplanation: {sample[2]}"
    return formatted_string

esnli_df.apply(get_inputs, axis=1)[2]
esnli_df.apply(get_response, axis=1)[2]
```

```python
from datasets import Dataset, concatenate_datasets

LIMIT = 5000

df_list = [gsm8k_df, aqua_df, creak_df, ecqa_df, esnli_df, qasc_df, qed_df, sensemaking_df, strategyqa_df]
limit_df_list = []
for df in df_list:
    if len(df) < LIMIT:
        limit_df_list.append(df)
    else:
        limit_df = df.sample(LIMIT, random_state=42).reset_index(drop=True)
        limit_df_list.append(limit_df)

merged_ds = [Dataset.from_pandas(df[['inputs', 'response']]) for df in limit_df_list]
dataset = concatenate_datasets(merged_ds)
dataset = dataset.shuffle(seed=42)
dataset
```
